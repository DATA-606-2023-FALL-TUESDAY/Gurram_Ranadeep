```python
def explore_cat_feature(df, feature):
    # Group by the specified feature and calculate the mean CO2 emissions
    group = df.groupby(feature)['CO2 Emissions(g/km)'].mean().sort_values()

    # Plot the mean CO2 emissions for each category
    plt.figure(figsize=[12, 6])
    sns.barplot(x=group.index, y=group.values, palette='viridis')
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Mean CO2 Emission', fontsize=12)
    plt.title(f"Mean CO2 Emission by {feature}\n", fontsize=15, fontweight='bold')
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

# List of categorical features to explore
cat_features = ['Make', 'Vehicle Class', 'Engine Size(L)', 'Cylinders', 'Transmission', 'Fuel Type']

# Call the function for each categorical feature
for feature in cat_features:
    explore_cat_feature(df, feature)
```

```python
def explore_num_feature(feature):
    plt.figure(figsize=[15, 5])
    plt.scatter(df[feature], df['CO2 Emissions(g/km)'])
    plt.xlabel(feature, fontsize=15)
    plt.ylabel('CO2 Emission', fontsize=15)
    plt.title(f"CO2 Emission vs. {feature}\n", fontsize=20)

# List of numerical features to explore
num_features = ['Engine Size(L)', 'Cylinders', 'Fuel Consumption Hwy (L/100 km)']

# Loop through the numerical features and call the function
for feature in num_features:
    explore_num_feature(feature)

# If you want to include a bar plot as in your code, you can do it separately:
plt.figure(figsize=[15, 5])
plt.bar(df.index, df["Fuel Consumption Hwy (L/100 km)"])
plt.xlabel('Index', fontsize=15)
plt.ylabel('Fuel Consumption Hwy (L/100 km)', fontsize=15)
plt.title("Fuel Consumption Hwy (L/100 km)\n", fontsize=20)
plt.show()
```
